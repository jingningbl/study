```python
import cv2
cap=cv2.VideoCapture(0)
while(True):
    ret,frame=cap.read()
    cv2.imshow(u"Capture",frame)
    key=cv2.waitKey(1)
    if key & 0xff==ord('q') or key ==27:
        print(frame.shape,ret)
        break
cap.release()
cv2.destroyAllWindows()
```
